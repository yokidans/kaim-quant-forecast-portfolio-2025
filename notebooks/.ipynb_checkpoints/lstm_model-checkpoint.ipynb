# LSTM Price Forecasting — Production Notebook

**Project:** KAIM Quant Forecast Portfolio  
**Notebook:** `lstm_model.ipynb`  
**Purpose:** Train and evaluate a robust LSTM model on processed price data (`Adj Close`) with professional diagnostics and artifact saving.

### What this notebook covers 
1. **Deterministic setup** (reproducibility, clean logging, directory hygiene)  
2. **Leakage-safe scaling & splits** (train/val/test by time)  
3. **Windowed time-series dataset builder** (lookback → next-step)  
4. **Modern Keras model** (Input layer, `.keras` format, callbacks: EarlyStopping, ReduceLROnPlateau, ModelCheckpoint)  
5. **Metrics** (MAE, RMSE, MAPE) with test-set only  
6. **Visual diagnostics**: 
   - Learning curves (train vs val loss)  
   - Actual vs Predicted (de-normalized)  
   - Residuals over time & histogram  
   - Residual autocorrelation (ACF)  
7. **Artifacts saved**: best model (`.keras`), predictions CSV, metrics JSON, plots PNG.

> **Data expectation**: Use files produced by your pipeline, e.g. `data/processed/TSLA_processed.csv` from `src/data/preprocess.py`, providing an `Adj Close` column indexed by date.


```python
# %% [setup] Imports, logging, reproducibility, folders
import os, json, math, random, time, warnings
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from statsmodels.tsa.stattools import acf
from packaging import version

plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.grid'] = True
warnings.filterwarnings('ignore')

# Make runs reproducible (best-effort)
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Create folders
PROJECT_ROOT = Path(r"C:\Users\tefer\kaim-quant-forecast-portfolio-2025")
DATA_DIR = PROJECT_ROOT / "data" / "processed"

TICKER = "TSLA"
DATA_PATH = DATA_DIR / f"{TICKER}_processed.csv"
MODELS_DIR = Path('models'); MODELS_DIR.mkdir(parents=True, exist_ok=True)
REPORTS_DIR = Path('reports'); REPORTS_DIR.mkdir(parents=True, exist_ok=True)
ARTIFACTS_DIR = Path('artifacts'); ARTIFACTS_DIR.mkdir(parents=True, exist_ok=True)

print('TensorFlow:', tf.__version__)
print('NumPy:', np.__version__)
print('Pandas:', pd.__version__)
```

```python
# %% [config] User-configurable parameters
TICKER = 'TSLA'            # or 'SPY', 'BND'
LOOKBACK = 60              # days to look back
HORIZON = 1                # forecast horizon (1 = next day)
EPOCHS = 50
BATCH_SIZE = 32
VAL_FRAC = 0.10            # 10% validation (after train split)
TEST_FRAC = 0.20           # 20% final test

# File auto-detection (works with {ticker}_processed.csv or {ticker}.csv)
CANDIDATES = [DATA_DIR / f"{TICKER}_processed.csv"]
for c in CANDIDATES:
    if c.exists():
        DATA_PATH = c
        break
else:
    raise FileNotFoundError(f"Could not find processed CSV for {TICKER} in {DATA_DIR}")
DATA_PATH
```

```python
# %% [data] Load series and basic sanity checks
df = pd.read_csv(DATA_PATH)

# Flexible column handling: if 'Adj Close' not present, try 'Close' or 'Price'
col_options = ['Adj Close', 'Close', 'Price']
target_col = next((c for c in col_options if c in df.columns), None)
if target_col is None:
    raise ValueError(f"None of {col_options} found in columns: {list(df.columns)}")

# Ensure date index if available
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df = df.dropna(subset=['Date']).set_index('Date').sort_index()

series = df[target_col].astype(float).values.reshape(-1, 1)
print(f"Loaded {len(series)} rows for {TICKER}. Using column: {target_col}")
df.tail()
```

```python
# %% [split+scale] Time-based split and leakage-safe scaling
n = len(series)
test_size = int(n * TEST_FRAC)
trainval = series[: n - test_size]
test = series[n - test_size :]

# Validation is a slice from the end of trainval
val_size = int(len(trainval) * VAL_FRAC)
train = trainval[: len(trainval) - val_size]
val = trainval[len(trainval) - val_size :]

scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train)
val_scaled = scaler.transform(val)
test_scaled = scaler.transform(test)

print(f"Split sizes → train: {len(train)}, val: {len(val)}, test: {len(test)}")
```

```python
# %% [windowing] Build windowed datasets (lookback → horizon)
def build_windows(arr, lookback=60, horizon=1):
    X, y = [], []
    for i in range(lookback, len(arr) - horizon + 1):
        X.append(arr[i - lookback : i, 0])
        y.append(arr[i + horizon - 1, 0])
    X = np.array(X)
    y = np.array(y)
    return X.reshape((X.shape[0], X.shape[1], 1)), y

X_train, y_train = build_windows(train_scaled, LOOKBACK, HORIZON)
X_val, y_val = build_windows(np.vstack([train_scaled[-LOOKBACK:], val_scaled]), LOOKBACK, HORIZON)
X_test, y_test = build_windows(np.vstack([trainval[-LOOKBACK:], test_scaled]), LOOKBACK, HORIZON)

print('Shapes:')
print('  X_train', X_train.shape, 'y_train', y_train.shape)
print('  X_val  ', X_val.shape,   'y_val  ', y_val.shape)
print('  X_test ', X_test.shape,  'y_test ', y_test.shape)
```

```python
# %% [model] Build LSTM (modern Keras API)
def build_lstm_model(lookback):
    model = Sequential([
        Input(shape=(lookback, 1)),
        LSTM(64, return_sequences=True),
        Dropout(0.2),
        LSTM(64, return_sequences=False),
        Dropout(0.2),
        Dense(32, activation='linear'),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

model = build_lstm_model(LOOKBACK)
model.summary()
```

```python
# %% [train] Train with strong callbacks and save best model
best_path = MODELS_DIR / f"{TICKER}_best_lstm.keras"
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1),
    ModelCheckpoint(filepath=str(best_path), monitor='val_loss', save_best_only=True)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=2,
    callbacks=callbacks
)

final_path = MODELS_DIR / f"{TICKER}_final_lstm.keras"
model.save(final_path)
print('Saved best model →', best_path)
print('Saved final model →', final_path)
```

```python
# %% [eval] Evaluate on test set (inverse transform metrics)
pred_scaled = model.predict(X_test)
pred = scaler.inverse_transform(pred_scaled.reshape(-1, 1)).ravel()
y_true = scaler.inverse_transform(y_test.reshape(-1, 1)).ravel()

mae = mean_absolute_error(y_true, pred)
rmse = math.sqrt(mean_squared_error(y_true, pred))
mape = np.mean(np.abs((y_true - pred) / y_true)) * 100
metrics = {'mae': float(mae), 'rmse': float(rmse), 'mape': float(mape)}
print(metrics)

# Save metrics and predictions
with open(ARTIFACTS_DIR / f'{TICKER}_lstm_metrics.json', 'w') as f:
    json.dump(metrics, f, indent=2)

pred_df = pd.DataFrame({
    'date': df.index[-len(y_true):],
    'actual': y_true,
    'pred': pred,
    'residual': y_true - pred
})
pred_df.to_csv(ARTIFACTS_DIR / f'{TICKER}_lstm_predictions.csv', index=False)
pred_df.tail()
```

```python
# %% [viz] Learning curves
plt.figure()
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Learning Curves (MSE)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.savefig(ARTIFACTS_DIR / f'{TICKER}_learning_curves.png', dpi=150)
plt.show()
```

```python
# %% [viz] Actual vs Predicted (test set)
plt.figure()
plt.plot(pred_df['date'], pred_df['actual'], label='Actual')
plt.plot(pred_df['date'], pred_df['pred'], label='Predicted')
plt.title(f'{TICKER} — Actual vs Predicted (Test)')
plt.xlabel('Date')
plt.ylabel(target_col)
plt.legend()
plt.tight_layout()
plt.savefig(ARTIFACTS_DIR / f'{TICKER}_actual_vs_pred.png', dpi=150)
plt.show()
```

```python
# %% [viz] Residuals over time and histogram
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
axes[0].plot(pred_df['date'], pred_df['residual'])
axes[0].axhline(0, color='black', linewidth=1)
axes[0].set_title('Residuals over Time (Actual - Predicted)')
axes[0].set_xlabel('Date'); axes[0].set_ylabel('Residual')

axes[1].hist(pred_df['residual'], bins=40)
axes[1].set_title('Residual Distribution (Histogram)')
axes[1].set_xlabel('Residual'); axes[1].set_ylabel('Frequency')
plt.tight_layout()
plt.savefig(ARTIFACTS_DIR / f'{TICKER}_residuals.png', dpi=150)
plt.show()
```

```python
# %% [viz] Residual ACF (autocorrelation)
res = pred_df['residual'].values
acf_vals = acf(res, fft=True, nlags=min(60, len(res)-2), missing='drop')
lags = np.arange(len(acf_vals))

plt.figure()
plt.stem(lags, acf_vals)  # Removed use_line_collection
plt.title('Residual Autocorrelation (ACF)')
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.tight_layout()
plt.savefig(ARTIFACTS_DIR / f'{TICKER}_residual_acf.png', dpi=150)
plt.show()

```
